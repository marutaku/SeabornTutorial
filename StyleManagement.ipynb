```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
np.random.seed(sum(map(ord, "aesthetics")))
```

```python
def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    #0から14までの区間を100等分する
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)
```

```python
sinplot()
# 通常のpyplot
```

```python
sns.set()
sinplot()
# seabornを適用することでスタイルが変化する
```

There are five preset seaborn themes: darkgrid, whitegrid, dark, white, and ticks.

```python
sns.set_style('whitegrid')
data = np.random.normal(size=(20, 6)) + np.arange(6) / 2
sns.boxplot(data=data);
```

データのパターンを見せたいだけの場合、グリッドは必要ない


```python
sns.set_style("dark")
sinplot()
```

```python
sns.set_style("white")
sinplot()
```

```python
sns.set_style("ticks")
sinplot()
```

グラフの線が出てくる

despineを使うことで上の枠と右の枠を消すことができる

```python
sinplot()
sns.despine()
```

目盛をあたいの始まるところまでカットすることもできる

```python
f, ax = plt.subplots()
sns.violinplot(data=data)
sns.despine(offset=5, trim=True);
```

Y軸を消すには、leftをTrueにする

```python
sns.set_style("whitegrid")
sns.boxplot(data=data, palette="deep")
sns.despine(left=True)
```

sns.axes_styleを用いることで、そのスコープの中だけでスタイルを変えることができる

```python
with sns.axes_style("darkgrid"):
    plt.subplot(211)
    sinplot()
plt.subplot(212)
sinplot(-1)
```

seabornのスタイルは、オーバーライドすることができる。  
オーバーライドをするときは、`axis_style`か`set_style`を使う  

パラメータの一覧を見るためには、引数を渡さずに実行すると現在のパラメータを返す

```python
sns.axes_style()
```

```python
sns.set_style("darkgrid", {"axes.facecolor": ".9"})
sinplot()
```

seabornがもともと用意してあるプリセットを用いるためには、`set`メソッドを用いる

最初に`set`を呼び出して初期化を行う

```python
sns.set()
```

プリセットには、paper, notebook, talk,  posterの四つがある

```python
sns.set_context("paper")
sinplot()
```

```python
sns.set_context("talk")
sinplot()
```

```python
sns.set_context("poster")
sinplot()
```

```python
sns.set_context("notebook", font_scale=1.5, rc={"lines.linewidth": 2.5})
sinplot()
```
